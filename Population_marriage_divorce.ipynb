```python
# Перегляд унікальних значень для ключових стовпців, щоб визначити яку інформацію (шум) можна прибрати і яку обробити
cols_to_check = ['Показник', 'Територіальний розріз', 'Вік', 'Стать', 'Тип місцевості',
       'Місяць реєстрації події', 'Сімейний cтан', 'Порядок розірвання шлюбу', 'Періодичність',
       'Період', 'Одиниця виміру', 'Масштабування',
       'Кількість десяткових знаків', 'Спеціальне значення даних',
       'Примітки набору даних', 'Примітки часового ряду',
       'Примітки спостереження']

for col in cols_to_check:
    print(f"🔹 {col}:")
    print(df[col].unique())
    print("-" * 30)
```

```python
# Переводимо стовпчик Місяць реєстрації та Період у формат дати щоб будувати часові ряди
# Словник українських місяців
month_map = {
    'Січень': '01', 'Лютий': '02', 'Березень': '03',
    'Квітень': '04', 'Травень': '05', 'Червень': '06',
    'Липень': '07', 'Серпень': '08', 'Вересень': '09',
    'Жовтень': '10', 'Листопад': '11', 'Грудень': '12',
    'Не застосовується': 'None'
}
```

```python
#  Замінюємо назви місяців на цифри
df['month_num'] = df['Місяць реєстрації події'].map(month_map)

# Створюємо строкову колонку формату "РІК-МІСЯЦЬ-01"
df['temp_date_str'] = df['Період'].astype(str) + '-' + df['month_num'] + '-01'

# Конвертуємо у формат datetime
df['Початок періоду'] = pd.to_datetime(df['temp_date_str'], errors='coerce')

# Видаляємо допоміжні колонки
df = df.drop(columns=['month_num', 'temp_date_str'])

df
```

```python
print(df['Початок періоду'].dtype)
```

Очищення датасету від "шумових" колонок, які не несуть аналітичної цінності

```python
# Список колонок для видалення
cols_to_remove = [
    'Періодичність',
    'Масштабування',
    'Кількість десяткових знаків',
     'Примітки набору даних', 'Примітки часового ряду', 'Примітки спостереження'
]

df = df.drop(columns=[c for c in cols_to_remove if c in df.columns])

# Перевіряємо результат
print("Залишилися колонки:", df.columns.tolist())
```

Також видаляємо рядки для стовпчику "Спеціальне значення даних" з атрибутом "Явищ не було"

```python
df = df[df['Спеціальне значення даних'] != 'Явищ не було']

# Після цього видаляємо сам стовпець, бо він став порожнім
df = df.drop(columns=['Спеціальне значення даних'])

print(f"Залишилося рядків: {len(df)}")
```

```python
# 1. Створюємо стовпчик "Кількість"
# Якщо Одиниця виміру — це Особа АБО Одиниця, копіюємо число, інакше — порожньо (NaN)
df['Кількість'] = np.where(
    df['Одиниця виміру'].isin(['Особа', 'Одиниця']),
    df['Значення cпостереження'],
    np.nan
)

# 2. Створюємо стовпчик "Вік_показник"
# Якщо Одиниця виміру — це Рік, копіюємо число
df['Вік_показник'] = np.where(
    df['Одиниця виміру'] == 'Рік',
    df['Значення cпостереження'],
    np.nan
)

# 3. Створюємо стовпчик "Коефіцієнт на 1000 осіб"
# Якщо Одиниця виміру — це відповідний текст, копіюємо число
df['Коефіцієнт на 1000 осіб'] = np.where(
    df['Одиниця виміру'] == 'На 1000 осіб наявного населення',
    df['Значення cпостереження'],
    np.nan
)

# Перевіряємо результат
print(df[['Одиниця виміру', 'Значення cпостереження', 'Кількість', 'Вік_показник', 'Коефіцієнт на 1000 осіб']].head(20))





```
